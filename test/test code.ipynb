```python
import requests, json 
from openpyxl import load_workbook

url = 'http://localhost:3000/rpa/upload'

load_excel = load_workbook('./example.xlsx', data_only=True)
load_sheet = load_excel['Sheet1']


conditional = True
index = 2

while(conditional):
    if (load_sheet[f'E{index}'].value == None):
        conditional = False

    data = {
        'title' : load_sheet[f'E{index}'].value
    }
    
    print(data)

    index += 1
    res = requests.post(url, json=data).json()



print('완료')
```
